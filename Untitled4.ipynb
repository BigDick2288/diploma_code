```python
def papapa(df):
    Total_Liabilities = df.iloc[:, :12]
    Total_Liabilities['FY-7'] = np.nan
    Total_Assets = df.iloc[:, 12:19]
    Total_Cash = df.iloc[:, 19:26]
    Cost_of_Debt = df.iloc[:, 26:33]
    Cost_of_Equity = df.iloc[:, 33:40]
    EBITDA = df.iloc[:, 40:47]
    Revenue_Business = df.iloc[:, 47:55]
    Market_Cap = df.iloc[:, 55:62]
    Current_Assets_Percentage = df.iloc[:, 62:69]
    Int_Paid = df.iloc[:, 69:76]

    
    df_name = [Total_Assets, Total_Cash, Cost_of_Debt, Cost_of_Equity, EBITDA, 
            Revenue_Business, Market_Cap, Current_Assets_Percentage, Int_Paid]
    col_name = ['Total_Assets', 'Total_Cash', 'Cost_of_Debt', 'Cost_of_Equity', 'EBITDA', 
            'Revenue_Business', 'Market_Cap', 'Current_Assets_Percentage', 'Int_Paid']
    t_l=Total_Liabilities.melt(col_level = 0, 
                       id_vars = ['Identifier (RIC)', 'Company Name', 'Country of Headquarters',
       'Company Market Cap\n(USD)', 'TRBC Economic Sector Name'], 
                       value_vars = ['FY0', 'FY-1', 'FY-2', 'FY-3', 'FY-4', 'FY-5', 'FY-6', 'FY-7'],
                       var_name = 'Year', 
                       value_name = 'Total Liabilities(USD)In the last 7 FY')
    a = []
    a.append(t_l)
    
    for i in range(len(df_name)):
        name = col_name[i]
        df_name[i]['Identifier (RIC)'] = df['Identifier (RIC)']

        if i!=5 : 
            df_name[i]['FY-7'] = np.nan
            df_name[i].columns = ['FY0', 'FY-1', 'FY-2', 'FY-3', 'FY-4', 'FY-5', 'FY-6', 'Identifier (RIC)', 'FY-7']
            r_a= df_name[i].melt(col_level = 0, 
                                 id_vars = ['Identifier (RIC)'], 
                                 value_vars = ['FY0', 'FY-1', 'FY-2', 'FY-3', 'FY-4', 'FY-5', 'FY-6', 'FY-7'],
                                 var_name = 'Year', 
                                 value_name = f'{name} In the last 7 FY')
        else: 
            df_name[i].columns = ['FY0', 'FY-1', 'FY-2', 'FY-3', 'FY-4', 'FY-5', 'FY-6', 'FY-7', 'Identifier (RIC)']
            r_a= df_name[i].melt(col_level = 0, 
                                 id_vars = ['Identifier (RIC)'], 
                                 value_vars = ['FY0', 'FY-1', 'FY-2', 'FY-3', 'FY-4', 'FY-5', 'FY-6', 'FY-7'],
                                 var_name = 'Year', 
                                 value_name = f'{name}In the last 8 FY')
        a.append(r_a)
    
    merged_df = reduce(lambda l, r: pd.merge(l, r, on=['Identifier (RIC)', 'Year']), a)
    return merged_df
```

```python


def clean_data(df): 
  df = df.dropna(subset = ['Total_Assets In the last 7 FY', 'Total Liabilities(USD)In the last 7 FY'])
  df = df[df['Total Liabilities(USD)In the last 7 FY'] != 0]
  df['D/A'] = df['Total Liabilities(USD)In the last 7 FY']/df['Total_Assets In the last 7 FY']
  df['cost_of_debt'] = df['Int_Paid In the last 7 FY']/df['Total Liabilities(USD)In the last 7 FY']
  df['ND/A'] = (df['Int_Paid In the last 7 FY'] -df['Total_Cash In the last 7 FY'] )/df['Total Liabilities(USD)In the last 7 FY']
  df = df[df['cost_of_debt']< 0.6]
  df = df[df['D/A'] < 1]
  df = df[df['ND/A'] < 1]
  df = df[df['ND/A'] > -0.25]
  df = df[df['cost_of_debt'] > 0]

  return df






```

```python

def changes(df):
  condition_b = [(df['Year'] == 'FY0'), (df['Year'] == 'FY-1'), (df['Year'] == 'FY-2'), (df['Year'] == 'FY-3'), (df['Year'] == 'FY-4'), (df['Year'] == 'FY-5'), (df['Year'] == 'FY-6'), (df['Year'] == 'FY-7')]
  values_с = [0.0151, 0.0093, 0.0192, 0.0268, 0.0241, 0.0245, 0.0227, 0.0217]
  values_d = [0.0424, 0.0472, 0.052, 0.0596, 0.0508, 0.0569, 0.0612, 0.0578]
  values_b = [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014]
  df['risk_free'] = np.select(condition_b, values_с)
  df['ERP'] = np.select(condition_b, values_d)
  df['Year'] = np.select(condition_b, values_b)
  return df




```

```python
def cals_beta (st, i):
    df = data[['^GSPC', str(st)]].dropna()

    if i < 60: 
        return np.NaN
    elif i > 60: 
        i = int(i)
        X = df[['^GSPC']].pct_change().iloc[i-60:i].to_numpy().reshape(-1, 1)
        Y = df[[st]].pct_change().iloc[i-60:i].to_numpy().reshape(-1, 1)
        if (len(X) < 5) or (len(Y) < 5): 
          return np.NaN
        else:
          lin_regr = LinearRegression()
          lin_regr.fit(X, Y)

          return lin_regr.coef_[0, 0]
```

```python
def wide_to_long(df):
  pan_beta = pd.melt(df, id_vars='Date', value_vars=l)
  pan_beta['Year'] = pan_beta['Date'].astype('datetime64[Y]')
  pan_beta = pan_beta.groupby(["Year", 'variable']).mean()
  pan_beta = pan_beta.reset_index()
  pan_beta = pan_beta.rename(columns={"variable": "Identifier (RIC)"})
  pan_beta = pan_beta.rename(columns={"value": "beta"})
  return pan_beta
```
